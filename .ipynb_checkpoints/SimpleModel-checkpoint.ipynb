```python
from fastai.vision.all import *
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
path = Path("19")
```

```python
path
```

```python
IMG_SIZE = 256
BATCH_SIZE = 32

# Autoencoder: target is the same as input
dblock = DataBlock(
    blocks=(ImageBlock, ImageBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=lambda x: x,   # target = input
    item_tfms=Resize(IMG_SIZE)
)

dls = dblock.dataloaders(path, bs=BATCH_SIZE)
```

```python
import torch.nn as nn

class AutoEncoder(nn.Module):
    def __init__(self):
        super().__init__()

        # Encoder: 128 → 64 → 32
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 16, 3, stride=2, padding=1),   # 128 → 64
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),  # 64 → 32
            nn.ReLU()
        )

        # Decoder: 32 → 64 → 128
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(16, 3, 3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.decoder(self.encoder(x))
```

```python
learn = Learner(dls, AutoEncoder(), loss_func=MSELossFlat())
```

```python
learn.fit(5)
```

```python
# Get a batch from validation set
xb, _ = dls.valid.one_batch()

# Move batch to same device as the dataloader/model
device = learn.dls.device
xb = xb.to(device)

# Forward pass
preds = learn.model(xb)

# Move back to CPU for visualization
preds = preds.cpu()
xb = xb.cpu()
```

```python
n = 6  # number of images to show
plt.figure(figsize=(12,4))
for i in range(n):
    # Original
    plt.subplot(2,n,i+1)
    plt.imshow(xb[i].permute(1,2,0))
    plt.title("Original")
    plt.axis('off')
    # Reconstructed
    plt.subplot(2,n,n+i+1)
    plt.imshow(preds[i].permute(1,2,0).detach())
    plt.title("Reconstructed")
    plt.axis('off')
plt.show()
```

```python
reconstruction_errors = ((preds - xb)**2).mean(dim=[1,2,3])
print(reconstruction_errors)
```
